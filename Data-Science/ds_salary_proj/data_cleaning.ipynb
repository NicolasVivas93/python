```python
import pandas as pd

df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df
```

```python
# Limpiar valores nulos / negativos (-1)
df = df[df['Salary Estimate'] != '-1']
df
```

```python
# Quitar 'Glassdoor est.'
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

```python
# Quitar los 'K' y signos de $ 
minusKd = salary.apply(lambda x: x.replace('K', '').replace('$',''))
minusKd
```

```python
df['Per hour salary'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df
```

```python
# Quitar texto 'Per hour' de salary_estimate
min_hr = minusKd.apply(lambda x: x.lower().replace('per hour', ''))
min_hr
```

```python
# Crear columna para mínimo de salary
df['Min_salary'] = min_hr.apply(lambda x: float(x.split('-')[0]))
min_hr.values[0] = '29.00 - 29.00'
min_hr.values[0]

# Crear columna para máximo de salary
df['Max_salary'] = min_hr.apply(lambda x: float(x.split('-')[1]))
df
```

```python
# Crear columna para el promedio de salario
df['Avg Salary'] = (df['Min_salary'] + df['Max_salary']) / 2
df
```

```python
# Separar state en un campo distinto

df.Location = df.Location.apply(lambda x: 'Remote,Remote' if x == 'Remote' else x)
df['State'] = df['Location'].apply(lambda x: x.split(',')[1])
df.Location = df.Location.apply(lambda x: x.split(',')[0])
df
```

```python
fifth_column = df.pop('State')
df.insert(5,'State', fifth_column)
df 
```

```python
#Contar cantidad de jobs para cada state
df.State.value_counts()
```

```python
df.to_csv('salary_clean.csv', index=False)
```
